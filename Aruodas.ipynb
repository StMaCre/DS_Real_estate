```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("/content/file.csv")
```

```python
print(df)
```

```python
df.describe()
```

# Cleaning the data

```python
list_clean = []
for row in df["Price"]:
  list_clean.append(row[:-5])

good_list = []
for element in list_clean: 
  good_list.append(int(element.replace(" ", "")))
df["good_price"] = good_list
```

```python
sep = '\n'
list_region = []
for row in df["Region"]: 
  list_region.append(row.split(sep, 1)[0])
print(list_region)

df["region_good"] = list_region
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
print(df)
```

```python
df.describe()
```

# How many new houses are available per region?



```python
houses_region = df["region_good"].value_counts()
print(houses_region)
```

# What is the mean price per region?

```python
from matplotlib import pyplot as plt
```

```python
grouped_region = df.groupby(['region_good'])['good_price'].mean()
region_sorted_price = grouped_region.sort_values(ascending=False)
print(region_sorted_price)
```

```python
ax = region_sorted_price.plot.bar(figsize=(18,5))
plt.title('Average price per square meter per region')
plt.show()
```

```python
ax = houses_region.plot.bar(figsize=(18,5))
plt.title('Number of houses available per region ')
plt.show()
```

# Availability of the houses / mean price

```python
# Create the DataFrame
frame = { 'mean_price': region_sorted_price, 'number_houses': houses_region}
availability = pd.DataFrame(frame)
avail_sorted = availability.sort_values("mean_price", ascending=False)
print(avail_sorted)
```

```python
# Create a scatter plot with region names
fig, ax = plt.subplots(figsize=(18,15))
ax.scatter(avail_sorted["mean_price"], avail_sorted["number_houses"])

for i, region in enumerate(avail_sorted.index):
    ax.annotate(region, (avail_sorted.loc[region, "mean_price"], avail_sorted.loc[region, "number_houses"]))

# Set axis labels and title
ax.set_xlabel("Average price")
ax.set_ylabel("Number of houses")
ax.set_title("Availability by region")

# Show the plot
plt.show()
```

# Distribution of prices per region

```python
# Group the data by region_good and create a list of lists containing the prices for each region
prices_by_region = [list(region_df['good_price']) for _, region_df in df.groupby('region_good')]

# Get the list of unique region names
region_labels = df['region_good'].unique()

# Create a box plot of the prices by region
fig, ax = plt.subplots(figsize=(18, 20))
ax.boxplot(prices_by_region, labels=region_labels, vert=False)

# Set axis labels and title
ax.set_ylabel('Box plot of Prices by Region')
ax.set_title('Prices')

plt.show()
```

# Distribution of the property size per region

```python
# Group the data by region_good and create a list of lists containing the property sizes for each region
sizes_by_region = [region_df['Ares'].tolist() for _, region_df in df.groupby('region_good')]

# Get the list of unique region names
region_labels = df['region_good'].unique()

# Create a box plot of the property sizes by region
fig, ax = plt.subplots(figsize=(10, 15))
ax.boxplot(sizes_by_region, labels=region_labels, vert=False)

# Set axis labels and title
ax.set_xlabel('Property Size (m²)')
ax.set_title('Box plot of Property Sizes by Region')

# Rotate the y-axis labels for better readability
ax.tick_params(axis='y', labelrotation=0)

plt.show()
```
